```python
# Import the os library to interact with the file system.
import os

# Import the Image module from the PIL (Pillow) library for image manipulation.
from PIL import Image

# Define the target size for the images.
image_size = (256, 256)

# Define the paths for the input and output directories.
input_dir = "/kaggle/input/pre-process-data/For-pre-process"
output_dir = "/kaggle/working/resized-data"

# Ensure the main output directory exists.
os.makedirs(output_dir, exist_ok=True)

# Loop through each class folder in the input directory.
for class_name in os.listdir(input_dir):
    # Construct the full path to the class folder.
    class_path = os.path.join(input_dir, class_name)
    
    # Create the corresponding output directory for the current class.
    os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

    # Loop through each image file in the current class folder.
    for image_name in os.listdir(class_path):
        # Construct the full path for both the original and new images.
        input_image_path = os.path.join(class_path, image_name)
        output_image_path = os.path.join(output_dir, class_name, image_name)
        
        try:
            # Open the image, resize it, and save it to the output path.
            Image.open(input_image_path).resize(image_size, Image.LANCZOS).save(output_image_path)
        except Exception as e:
            # Handle any potential errors gracefully.
            print(f"Could not process {image_name}: {e}")

# This message will only print once after all images have been processed.
print("\nAll images resized and saved successfully! ")
```

```python
# Import necessary libraries for data manipulation and visualization.
import matplotlib.pyplot as plt
import seaborn as sns
import os
from PIL import Image
```

```python
# Define paths to your original and resized datasets.
original_data_path = "/kaggle/input/pre-process-data/For-pre-process"
resized_data_path = "/kaggle/working/resized-data"

# Lists to store the dimensions (width and height) of images.
original_dimensions = []
resized_dimensions = []

# Function to collect dimensions from a given directory.
def collect_dimensions(directory, dimensions_list):
    # Loop through each class folder.
    for class_name in os.listdir(directory):
        class_path = os.path.join(directory, class_name)
        # Check if the path is a directory to avoid errors.
        if os.path.isdir(class_path):
            # Loop through each image in the class folder.
            for image_name in os.listdir(class_path):
                image_path = os.path.join(class_path, image_name)
                # Ensure the file is a valid image.
                if image_name.endswith(('.jpg', '.jpeg', '.png')):
                    try:
                        with Image.open(image_path) as img:
                            # Get the width and height of the image and append it to the list.
                            dimensions_list.append(img.size)
                    except Exception as e:
                        print(f"Skipping {image_name} due to an error: {e}")

# Collect dimensions from the original dataset.
collect_dimensions(original_data_path, original_dimensions)
# Collect dimensions from the resized dataset.
collect_dimensions(resized_data_path, resized_dimensions)

# Now, let's create the plots.
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot for the original images.
original_widths = [dim[0] for dim in original_dimensions]
original_heights = [dim[1] for dim in original_dimensions]
sns.histplot(x=original_widths, y=original_heights, ax=axes[0], color="skyblue", cbar=True)
axes[0].set_title('Original Image Dimensions (Before Pre-processing)')
axes[0].set_xlabel('Width')
axes[0].set_ylabel('Height')

# Plot for the resized images.
resized_widths = [dim[0] for dim in resized_dimensions]
resized_heights = [dim[1] for dim in resized_dimensions]
sns.histplot(x=resized_widths, y=resized_heights, ax=axes[1], color="coral", cbar=True)
axes[1].set_title('Resized Image Dimensions (After Pre-processing)')
axes[1].set_xlabel('Width')
axes[1].set_ylabel('Height')

# Show the plots.
plt.tight_layout()
plt.show()

print("\nImage dimension visualization complete! Check the plots above. ðŸ“ˆ")
```

```python
# Function to get pixel data from a few sample images.
def get_pixel_data(directory, num_samples=3):
    pixel_data = {}
    class_names = os.listdir(directory)
    
    # Iterate through the first few class folders to get samples.
    for class_name in class_names[:num_samples]:
        class_path = os.path.join(directory, class_name)
        if os.path.isdir(class_path):
            image_files = [f for f in os.listdir(class_path) if f.endswith(('.jpg', '.jpeg', '.png'))]
            # Get the first image from each of the selected classes.
            if image_files:
                sample_image_path = os.path.join(class_path, image_files[0])
                with Image.open(sample_image_path) as img:
                    # Convert to grayscale to get a single channel for the histogram.
                    grayscale_img = img.convert('L')
                    # Flatten the image data into a single list of pixel values.
                    pixel_data[f"{class_name} ({os.path.basename(directory)})"] = list(grayscale_img.getdata())
    return pixel_data

# Collect pixel data from both datasets.
original_pixel_data = get_pixel_data(original_data_path)
resized_pixel_data = get_pixel_data(resized_data_path)

# Create a figure for the plots.
plt.figure(figsize=(15, 6))

# Plot the histograms for original images.
for i, (label, data) in enumerate(original_pixel_data.items()):
    plt.subplot(1, 2, 1) # Plot on the first subplot
    sns.histplot(data, bins=50, kde=True, label=label)

# Set title and labels for the first plot.
plt.title('Pixel Intensity Distribution (Original Images)')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.legend()


# Plot the histograms for resized images.
for i, (label, data) in enumerate(resized_pixel_data.items()):
    plt.subplot(1, 2, 2) # Plot on the second subplot
    # Adjust label for the resized data for clarity.
    new_label = label.replace("For-pre-process", "Resized-data")
    sns.histplot(data, bins=50, kde=True, label=new_label)

# Set title and labels for the second plot.
plt.title('Pixel Intensity Distribution (Resized Images)')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.legend()

# Show the plots.
plt.tight_layout()
plt.show()

print("\nPixel intensity visualization complete! ")
```
